```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import pywt
from statsmodels.robust import mad
import tisean
%matplotlib inline
import denoise_functions.Denoise_Functions as df
import denoise_functions.Filters as ft
```

```python
temp = np.linspace(-10, 10, 20001)
func = df.Denoise_Functions(temp)
f1_plot = func.f1_plot()
f2_plot = func.f2_plot()
f3_plot = func.f3_plot()
f4_plot = func.f4_plot()
f5_plot = func.f5_plot()

plt.figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(2, 3, 1)
plt.plot(temp, f1_plot, label='function 1', linewidth=1)
plt.title('Function #1')
plt.legend()
plt.subplot(2, 3, 2)
plt.plot(temp, f2_plot, label='function 2', linewidth=1)
plt.title('Function #2')
plt.legend()
plt.subplot(2, 3, 3)
plt.plot(temp, f3_plot, label='function 3', linewidth=1)
plt.title('Function #3')
plt.legend()
plt.subplot(2, 3, 4)
plt.plot(temp, f4_plot, label='function 4', linewidth=1)
plt.title('Function #4')
plt.legend()
plt.subplot(2, 3, 5)
plt.plot(temp, f5_plot, label='function 5', linewidth=1)
plt.title('Function #5')
plt.legend()
```

```python
f1 = func.f1()
series = f1 + np.random.randn(len(temp))*0.1
filt = ft.Filters(series)
ma = filt.moving_average_filter()
em = filt.exponential_smoothing_filter(a=0.2)
lf = filt.linear_fourier_smoothing(cutoff=5.0)
ws = filt.wavelet_shrinkage(filt='db8')
ts = filt.tisean_denoise(embedding_dimension=8, iteration=1, neighbourhood_size=0.1, delay_time=314)

idx = np.where((temp>=-0.5) & (temp <= 0.5))
f1_part = f1[idx]
temp_part = temp[idx]
series_part = series[idx]
ma_part = ma[idx]
em_part = em[idx]
lf_part = lf[idx]
ws_part = ws[idx]
ts_part = ts[idx]

plt.figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(2, 3, 1)
plt.plot(temp_part, f1_part, label='original', linewidth=1)
plt.title('Original')
plt.legend()
plt.subplot(2, 3, 2)
plt.scatter(temp_part, series_part, color='r', label='with noise', s=0.5)
plt.plot(temp_part, ma_part, label='after smoothing', linewidth=1)
plt.title('Moving Average')
plt.legend()
plt.subplot(2, 3, 3)
plt.scatter(temp_part, series_part, color='r', label='with noise', s=0.5)
plt.plot(temp_part, ws_part, label='after smoothing', linewidth=1)
plt.title('Wavelet Shrinkage')
plt.legend()
plt.subplot(2, 3, 4)
plt.scatter(temp_part, series_part, color='r', label='with noise', s=0.5)
plt.plot(temp_part, ts_part, label='after smoothing', linewidth=1)
plt.title('TISEAN')
plt.legend()
plt.subplot(2, 3, 5)
plt.scatter(temp_part, series_part, color='r', label='with noise', s=0.5)
plt.plot(temp_part, em_part, label='after smoothing', linewidth=1)
plt.title('Exponential Smoothing')
plt.legend()
plt.subplot(2, 3, 6)
plt.scatter(temp_part, series_part, color='r', label='with noise', s=0.5)
plt.plot(temp_part, lf_part, label='after smoothing', linewidth=1)
plt.title('Linear Fourier Smoothing')
plt.legend()
```
